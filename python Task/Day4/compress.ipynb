```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# load an image 

image = Image.open("example.jpg")
Image_array = np.array(image)/ 255.0  #Normalize pixel value
Image_array.shape
```

```python
# Reshape the image to 2D (pixels x RGB)
h, w, d = Image_array.shape
Image_2D = Image_array.reshape(h*w,d)
Image_2D.shape
```

```python
kmeans = KMeans(n_clusters=16,random_state=0)
kmeans.fit(Image_2D)
compressed_colors = kmeans.cluster_centers_[kmeans.labels_]
compressed_image = compressed_colors.reshape(h,w,d)
print("Shape of compressed image = ",compressed_image.shape)

#Display and save the compressed image 
plt.imshow(compressed_image)
plt.axis("off")
plt.savefig('compressed.jpg', format='jpg')
```
