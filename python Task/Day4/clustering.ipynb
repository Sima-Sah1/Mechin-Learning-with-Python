```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans 
```

```python
X,y = make_blobs(n_samples=500,centers=3,cluster_std=0.60, random_state=0)
# Visualize the data 
plt.scatter(X[:,0],X[:,1], s=30)
plt.title("Generated Data")
plt.show()
```

```python
# Initialize KMeans with k
Kmeans = KMeans(n_clusters=2)

#Fit the model to the data
Kmeans.fit(X)

#Predict the clusters
y_Kmeans = Kmeans.predict(X)
# Visualize the clusters
plt.scatter(X[:,0],X[:,1],c=y_Kmeans, s=30,cmap='viridis')
centers = Kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='red',s=200,alpha=0.7,marker='X')
plt.title("KMeans")
plt.show()
```

```python
# Elbow Method - Compute WCSS(Within Cluster Sun of Squares) fr diffetent k 

wcss_list = []
for k in range(1,11):
    Kmeans = KMeans(n_clusters=k)
    Kmeans.fit(X)
    wcss_list.append(Kmeans.inertia_)


#Plot the elbow curve
plt.plot(range(1,11),wcss_list,marker='o')
plt.xlabel("k")
plt.ylabel("WCSS")
plt.title("k vs WCSS")
plt.show()    
```

```python

from sklearn.datasets import make_circles
X,y = make_circles(n_samples=500,factor=0.5,noise=0.05,random_state=42)
plt.scatter(X[:,0],X[:,1], s=30)
plt.title("Generated Data")
plt.show()
```

```python
# Initialize KMeans with k
Kmeans = KMeans(n_clusters=2)

#Fit the model to the data
Kmeans.fit(X)

#Predict the clusters
y_Kmeans = Kmeans.predict(X)
# Visualize the clusters
plt.scatter(X[:,0],X[:,1],c=y_Kmeans, s=30,cmap='viridis')
centers = Kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='red',s=200,alpha=0.05,marker='X')
plt.title("KMeans")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
X,y = make_circles(n_samples=500,factor=0.5,noise=0.05,random_state=42)
plt.scatter(X[:,0],X[:,1], s=30)
plt.title("Generated Data")
plt.show()
```

```python
dbscan = DBSCAN(eps= 0.175,min_samples=10)
y_dbscan =dbscan.fit_predict(X)
plt.scatter(X[:,0], X[:,1],c=y_dbscan,s=50,cmap='viridis')
plt.title("DBSCAN")
plt.show()
DBSCAN()
```

```python
from sklearn.neighbors import NearestNeighbors

#  Compute the K-distance plot
neighbors = NearestNeighbors(n_neighbors=10)
neighbors_fit = neighbors.fit(X)
distances,indices = neighbors_fit.kneighbors(X)

#sort and plot the distance
distances = np.sort(distances[:,9],axis=0)
plt.plot(distances)
plt.xlabel("Points")
plt.ylabel("Epsilon")
plt.title("K-distance Plot")
plt.show()
```
